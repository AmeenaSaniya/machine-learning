```python
classes = ['Class1.1', 'Class1.2', 'Class1.3', 'Class2.1', 'Class2.2', 'Class3.1','Class3.2', 'Class4.1', 'Class4.2', 'Class5.1', 'Class5.2', 'Class5.3','Class5.4', 'Class6.1', 'Class6.2', 'Class7.1', 'Class7.2', 'Class7.3','Class8.1', 'Class8.2', 'Class8.3', 'Class8.4', 'Class8.5', 'Class8.6', 'Class8.7', 'Class9.1', 'Class9.2', 'Class9.3', 'Class10.1', 'Class10.2', 'Class10.3', 'Class11.1', 'Class11.2', 'Class11.3', 'Class11.4', 'Class11.5', 'Class11.6'
]
```

```python
def append_ext(fn):
    return fn + ".jpg"
```

```python
import pandas as pd
```

```python
traindf = pd.read_csv('/content/drive/MyDrive/new/training_solutions_rev1.csv')
```

```python
traindf
```

```python
traindf["GalaxyID"] = traindf['GalaxyID'].astype(str).apply(append_ext)
```

```python
traindf
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(
    fill_mode='nearest',
    cval=0,
    rescale=1. / 255,
    rotation_range=90,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    validation_split=0.02)
```

```python
train_generator = datagen.flow_from_dataframe(
    dataframe=traindf,
    directory="/content/drive/MyDrive/new/saniya/train/images_training_rev1/images_training_rev1",
    x_col="GalaxyID",
    y_col=classes,
    subset="training",
    batch_size=64,
    seed=123,
    shuffle=True,
    class_mode="raw",
    target_size=(224, 224))
```

```python
valid_generator = datagen.flow_from_dataframe(
    dataframe=traindf,
    directory="/content/drive/MyDrive/new/saniya/train/images_training_rev1/images_training_rev1",
    x_col="GalaxyID",
    y_col=classes,
    subset="validation",
    batch_size=64,
    seed=123,
    shuffle=True,
    class_mode="raw",
    target_size=(224, 224))
```

```python
train_size = train_generator.n // train_generator.batch_size
valid_size = valid_generator.n // valid_generator.batch_size
```

```python
train_size,valid_size
```

```python
import keras
print(dir(keras.applications))
```

```python
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.preprocessing.image import load_img,img_to_array
from keras.models import Model
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_shape = (224, 224, 3)
vgg_model = VGG16(include_top=False, input_shape=img_shape)
```

```python
# Flatten output of last layer before adding output layer (Dense layer)
from keras.layers.core import Flatten, Dense, Dropout, Lambda, Reshape
x = Flatten()(vgg_model.output)
# Add output layer (number of outputs = 37)
x = Dense(len(classes), activation='sigmoid')(x)
# Load the modified model
model = Model(inputs=vgg_model.input, outputs=x)
```

```python
for layer in model.layers:
    layer.trainable = True
```

```python
optimizer = keras.optimizers.Adam(lr=0.001, decay=5e-4)
model.compile(optimizer, loss='mse', metrics=["accuracy"])
from keras.callbacks import Callback
from keras.callbacks import ModelCheckpoint, Callback, EarlyStopping
```

```python
from keras.callbacks import ModelCheckpoint
checkpointer = ModelCheckpoint(
    filepath='today_20_epochs/weights.hdf5', verbose=2, save_best_only=True)
```

```python
hist = model.fit_generator(
    train_generator,
    steps_per_epoch=train_size,
    validation_data=valid_generator,
    validation_steps=valid_size,
    epochs=15,
    #callbacks=[history, checkpointer, early_stopping])
    callbacks=[checkpointer])
```
